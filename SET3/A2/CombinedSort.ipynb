```python
import matplotlib.pyplot as plt
import numpy as np
import glob
```

```python
def read_results(file_path):
    sizes = []
    times = []
    with open(file_path, 'r') as f:
        for line in f:
            if line.strip():
                s, t = line.strip().split()
                sizes.append(int(s))
                times.append(float(t))
    return np.array(sizes), np.array(times)
```

```python
merge_files = {
    'random': 'merge_results_unordered1.txt',
    'back': 'merge_results_back_ordered1.txt',
    'almost': 'merge_results_almost_ordered1.txt'
}

combined_files = {
    5: {
        'random': 'combined5_results_unordered1.txt',
        'back': 'combined5_results_back_ordered1.txt',
        'almost': 'combined5_results_almost_ordered1.txt'
    },
    10: {
        'random': 'combined10_results_unordered1.txt',
        'back': 'combined10_results_back_ordered1.txt',
        'almost': 'combined10_results_almost_ordered1.txt'
    },
    20: {
        'random': 'combined20_results_unordered1.txt',
        'back': 'combined20_results_back_ordered1.txt',
        'almost': 'combined20_results_almost_ordered1.txt'
    },
    30: {
        'random': 'combined30_results_unordered1.txt',
        'back': 'combined30_results_back_ordered1.txt',
        'almost': 'combined30_results_almost_ordered1.txt'
    },
    50: {
        'random': 'combined50_results_unordered1.txt',
        'back': 'combined50_results_back_ordered1.txt',
        'almost': 'combined50_results_almost_ordered1.txt'
    }
}
```

```python
def plot_sort_results(array_type='random'):
    plt.figure(figsize=(12, 8))


    sizes, times = read_results(merge_files[array_type])
    plt.plot(sizes, times, label='MergeSort', linewidth=2)


    for barrier in sorted(combined_files.keys()):
        sizes, times = read_results(combined_files[barrier][array_type])
        plt.plot(sizes, times, label=f'CombinedSort barrier={barrier}')

    plt.xlabel('Размер массива', fontsize=14)
    plt.ylabel('Время (микросекунды)', fontsize=14)
    plt.title(f'Сравнение MergeSort и CombinedSort ({array_type} array)', fontsize=16)
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
for arr_type in ['random', 'back', 'almost']:
    plot_sort_results(arr_type)
```

```python
merge_files = {
    'random': 'merge_results_unordered1.txt',
    'back': 'merge_results_back_ordered1.txt',
    'almost': 'merge_results_almost_ordered1.txt'
}

combined_files = {
    5: {
        'random': 'combined5_results_unordered1.txt',
        'back': 'combined5_results_back_ordered1.txt',
        'almost': 'combined5_results_almost_ordered1.txt'
    },
    10: {
        'random': 'combined10_results_unordered1.txt',
        'back': 'combined10_results_back_ordered1.txt',
        'almost': 'combined10_results_almost_ordered1.txt'
    },
    20: {
        'random': 'combined20_results_unordered1.txt',
        'back': 'combined20_results_back_ordered1.txt',
        'almost': 'combined20_results_almost_ordered1.txt'
    },
    30: {
        'random': 'combined30_results_unordered1.txt',
        'back': 'combined30_results_back_ordered1.txt',
        'almost': 'combined30_results_almost_ordered1.txt'
    },
    50: {
        'random': 'combined50_results_unordered1.txt',
        'back': 'combined50_results_back_ordered1.txt',
        'almost': 'combined50_results_almost_ordered1.txt'
    }
}
```

```python
def plot_by_sort(sort_type='merge', barriers=None):
    plt.figure(figsize=(12, 8))

    if sort_type == 'merge':
        for arr_type, file_path in merge_files.items():
            sizes, times = read_results(file_path)
            plt.plot(sizes, times, label=f'{arr_type} array')
        plt.title('MergeSort: разные типы массивов', fontsize=16)
    else:
        for barrier in barriers:
            for arr_type in ['random', 'back', 'almost']:
                sizes, times = read_results(combined_files[barrier][arr_type])
                plt.plot(sizes, times, label=f'Barrier={barrier}, {arr_type}')
        plt.title('CombinedSort: разные типы массивов', fontsize=16)

    plt.xlabel('Размер массива', fontsize=14)
    plt.ylabel('Время (микросекунды)', fontsize=14)
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plot_by_sort('merge')
plot_by_sort('combined', barriers=[5, 10, 20, 30, 50])
```

```python
for barrier in combined_files.keys():
    plt.figure(figsize=(12, 8))
    for arr_type, file_path in combined_files[barrier].items():
        sizes, times = read_results(file_path)
        plt.plot(sizes, times, label=f'{arr_type} array')
    plt.title(f'CombinedSort with Barrier={barrier}', fontsize=16)
    plt.xlabel('Размер массива', fontsize=14)
    plt.ylabel('Время (микросекунды)', fontsize=14)
    plt.legend()
    plt.grid(True)
    plt.show()
```
