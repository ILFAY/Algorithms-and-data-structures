```python
import matplotlib.pyplot as plt
```

```python
quick_files = {
    "unordered": "quick_results_unordered1.txt",
    "back": "quick_results_back_ordered1.txt",
    "almost": "quick_results_almost_ordered1.txt"
}

intro_files = {
    "unordered": "intro_results_unordered1.txt",
    "back": "intro_results_back_ordered1.txt",
    "almost": "intro_results_almost_ordered1.txt"
}
```

```python
def read_data(filename):
    sizes = []
    times = []
    with open(filename, "r") as f:
        for line in f:
            s, t = line.strip().split()
            sizes.append(int(s))
            times.append(float(t))
    return sizes, times
```

```python
array_types = ["unordered", "back", "almost"]

for arr_type in array_types:
    q_sizes, q_times = read_data(quick_files[arr_type])
    i_sizes, i_times = read_data(intro_files[arr_type])

    plt.figure(figsize=(10,6))
    plt.plot(q_sizes, q_times, label="QuickSort", marker='o', markersize=2)
    plt.plot(i_sizes, i_times, label="IntroSort", marker='s', markersize=2)

    plt.title(f"QuickSort vs IntroSort ({arr_type} array)")
    plt.xlabel("Array size")
    plt.ylabel("Time (microseconds)")
    plt.grid(True)
    plt.legend()
    plt.show()
```
