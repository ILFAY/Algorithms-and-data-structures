```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
file_results = 'results.txt'
file_small = 'small_results.txt'
```

```python
data_results = np.loadtxt(file_results)
data_small = np.loadtxt(file_small)

N_results = data_results[:, 0]
s_results = data_results[:, 1]

N_small = data_small[:, 0]
s_small = data_small[:, 1]
```

```python
plt.figure(figsize=(12, 6))
plt.plot(N_results, s_results, lw=2, label='results')
plt.plot(N_small, s_small, lw=2, label='small results')
plt.xlabel(" ", fontsize=14)
plt.ylabel(" ", fontsize=14)
plt.title("    ", fontsize=16)
plt.grid(True)
plt.legend()
plt.show()
```

```python
s_small = np.abs(s_small - (0.25*np.pi + 1.25* np.arcsin(0.8) - 1))
s_results = np.abs(s_results - (0.25*np.pi + 1.25* np.arcsin(0.8) - 1))
```

```python
plt.figure(figsize=(12, 6))
plt.plot(N_results, s_results, lw=2, label='results')
plt.plot(N_small, s_small, lw=2, label='small results')
plt.xlabel(" ", fontsize=14)
plt.ylabel("   ", fontsize=14)
plt.title("        ", fontsize=16)
plt.grid(True)
plt.legend()
plt.show()
```
